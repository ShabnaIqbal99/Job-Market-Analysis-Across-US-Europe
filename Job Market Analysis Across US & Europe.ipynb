```python
# Contract time filled by words contained in 'title'

import re

def detect_contract_time(title):
    title = str(title).lower()
    
# Check percentage
    match = re.search(r"(\d{1,3})%", title)
    if match:
        percent = int(match.group(1))
        return "Full-time" if percent >= 90 else "Part-time"
    
# Using Keywords:
    if any(word in title for word in [
        "full time", "full-time", "vollzeit",
        "temps plein", "tiempo completo", "tempo pieno"
    ]):
        return "Full-time"
    
    if any(word in title for word in [
        "part time", "part-time", "teilzeit",
        "temps partiel", "tiempo parcial", "tempo parziale"
    ]):
        return "Part-time"
    
    return np.nan
```

```python
# Apply the above condition to Null Values of 'Contract time' only:

mask = df['contract_time'].isna()

df.loc[mask, 'contract_time'] = (
    df.loc[mask, 'title'].apply(detect_contract_time)
)
```

##### *There were few hints of contract time in the job title. Hence, filled the null value with the help of title that met the condition. Rest are kept as null*

```python
# Contract type filled by words contained in 'title'
def detect_contract_type(title):
    title = str(title).lower()
    
    if any(word in title for word in [
        "permanent", "unbefristet", "cdi"
    ]):
        return "permanent"
    
    if any(word in title for word in [
        "temporary", "fixed", "contract",
        "befristet", "zeitvertrag",
        "cdd", "interim", "freelance"
    ]):
        return "contract"
    
    return np.nan
```

```python
# Filled missing values in the salary_min, salary_mid, and salary_max columns by using the median salary for each combination of country and title. 

df['salary_min'] = df.groupby(
    ['country','title']
)['salary_min'].transform(lambda x: x.fillna(x.median()))

df['salary_mid'] = df.groupby(
    ['country','title']
)['salary_mid'].transform(lambda x: x.fillna(x.median()))

df['salary_max'] = df.groupby(
    ['country','title']
)['salary_max'].transform(lambda x: x.fillna(x.median()))
```

```python
# Filled missing values in the salary_min, salary_mid, and salary_max columns by using the median salary for each combination of country and category.

df['salary_min'] = df.groupby(
    ['country','category_label']
)['salary_min'].transform(lambda x: x.fillna(x.median()))

df['salary_mid'] = df.groupby(
    ['country','category_label']
)['salary_mid'].transform(lambda x: x.fillna(x.median()))

df['salary_max'] = df.groupby(
    ['country','category_label']
)['salary_max'].transform(lambda x: x.fillna(x.median()))
```

```python
# Filled missing values in the salary_min, salary_mid, and salary_max columns by using the median salary for each combination of country.

df['salary_min'] = df.groupby('country')['salary_min']\
                      .transform(lambda x: x.fillna(x.median()))

df['salary_mid'] = df.groupby('country')['salary_mid']\
                      .transform(lambda x: x.fillna(x.median()))

df['salary_max'] = df.groupby('country')['salary_max']\
                      .transform(lambda x: x.fillna(x.median()))
```

##### *Filled missing values in the salary_min, salary_mid, and salary_max columns by using the median salary for each combination of country and title,country and category, and with the country itself. This ensures that missing salaries are imputed based on similar job roles within the same country, preserving the overall distribution and reducing bias.*

#### Standardizing Text Columns

```python
# Propercase and strip string columns for consistency
str_cols = ['title', 'company_name', 'category_label', 
            'country', 'region','contract_type','contract_time']

for col in str_cols:
    df[col] = df[col].str.title().str.strip()
```

```python
import re

def clean_job_title(title):
    title = str(title).strip().lower()

    # Remove gender notations
    title = re.sub(r'\b(m|f|d|w|m/f/d|m/w/d|f/m/d|f/m/w)\b', '', title)
    title = re.sub(r'\((m|f|d|w|m/f/d|m/w/d)\)', '', title)

    # Remove percentages (with optional space before %)
    title = re.sub(r'\d{1,3}(\s*-\s*\d{1,3})?\s*%?', '', title)

    # Remove workload words like 'permanent', 'full-time', 'part-time' (optional)
    title = re.sub(r'\b(permanent|full[-\s]?time|part[-\s]?time|contract|temporary)\b', '', title)

    # Remove locations in parentheses
    title = re.sub(r'\([^)]*\)', '', title)

    # Remove locations after "in" or "at"
    title = re.sub(r'\b(in|at)\s+[a-z\s\-\'\d]+', '', title)

    # Remove extra punctuation and dashes
    title = re.sub(r'[\–\—\-\/|]', ' ', title)

    # Remove multiple spaces
    title = re.sub(r'\s+', ' ', title)

    # Strip again
    title = title.strip()

    # Optional: title-case each word
    title = title.title()

    return title

# Apply to dataframe
df['title'] = df['title'].apply(clean_job_title)
```

## Dataset Limitations

**1. Data Source Representation**
The dataset is collected from specific job listing sources, which may not fully represent the entire job market in each country. Some industries or smaller companies may be underrepresented.

**2. Incomplete Salary Information**
Not all job postings include salary details, and some salaries may be estimated or reported as ranges. This can affect the accuracy of salary trend analysis.

**3. Uneven Distribution of Data Across Years**
The dataset contains a significantly higher number of job listings in 2025 compared to previous years, which may reflect data availability rather than actual growth in job demand.

**4. Limited Regional Breakdown for Some Countries**
Certain countries only provide job listings at the national level without detailed regional information, limiting deeper geographic analysis.

**5. Broad Job Category Grouping**
Job roles are grouped into broad sector categories, which may oversimplify the diversity of job positions and skills within each industry.

## Future Improvements
